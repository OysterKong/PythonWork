```python
import json
import pandas as pd
import numpy as np
```

```python
with open('preprocessed_recipe_data.json', encoding="utf-8") as f:
    data = json.load(f)
```

```python
### 1. 재료 자체를 삭제

del_ingredient_list = ["육원전반죽", "무김치", "진육수", "육수", "어묵국물",
                        "장아찌국물", "통조림파인애플국물", "장조림국물", "오리백숙국물",
                        "감자채볶음", "아이스크림스쿱", "만두조림", "김칫국물", "백오이물김치국물", "깍두기김칫국물", "콩나물무침"]
for i in range(len(data['data'])):
    del_idx = []    # 삭제할 재료의 인덱스를 저장
    for c in range(len(data['data'][i]['ingredients'])):
        ingredient = data['data'][i]['ingredients'][c]

        # 리스트에 넣은 재료 삭제
        if(ingredient in del_ingredient_list):
            del_idx.append(c)

        # "딸기생크림케이크"에서 딸기 이후의 리스트 삭제
        if(data['data'][i]['title'] == "딸기생크림케이크"):
            if(ingredient == "딸기"):
                for k in range(c + 1, len(data['data'][i]['ingredients'])):
                    del_idx.append(k)

    del_idx.sort()
    # 저장된 인덱스의 재료를 삭제
    cnt = 0
    for idx in del_idx:
        idx = idx - cnt
        del(data['data'][i]['ingredients'][idx])
        # 앞에 재료가 하나 삭제 되면 인덱스가 하나씩 밀리게 되므로 처리
        cnt = cnt + 1
```

```python
###################################################################

### 2. 레시피를 삭제

del_receipe_by_ingredient_list = ["양념치킨", "찹스테이크샐러드의드레싱",
                                    "슈레드모차렐라치즈찹스테이크샐러드의스테이크와드레싱",
                                    "찹스테이크샐러드의스테이크와채소",
                                    "고추드레싱그린샐러드", "고추드레싱그린샐러드의채소"
                                    "고추드레싱그린샐러드의드레싱", 
                                    "맛살감자샐러드"]

del_recepe_idx_list = []
for i in range(len(data['data'])):
    if(data['data'][i]["title"] == "닭고기 비빔쌀국수"):
            del_recepe_idx_list.append(i)
    for c in range(len(data['data'][i]['ingredients'])):
        ingredient = data['data'][i]['ingredients'][c]

        # 리스트에 넣은 재료가 포함된 레시피의 인덱스 저장
        if(ingredient in del_receipe_by_ingredient_list):
            del_recepe_idx_list.append(i)

cnt = 0
for idx in del_recepe_idx_list:
    idx = idx - cnt
    del(data['data'][idx])
    # 앞에 재료가 하나 삭제 되면 인덱스가 하나씩 밀리게 되므로 처리
    cnt = cnt + 1                      
```

```python
###################################################################

### 3. 두개 이상의 재료가 합쳐져있는 재료 분할

mixed_ingredient_list = [["소금리코타치즈토스트재료호두", "소금/호두"],
                        ["미숫가루키위막걸리키위", "미숫가루/키위/막걸리"],
                        ["날치알슈레드모차렐라치즈", "날치알/슈레드모차렐라치즈"],
                        ["베이킹파우더달걀", "베이킹파우더/달걀"],
                        ["강력분베이킹파우더", "강력분/베이킹파우더"],
                        ["초코크런치와스프링클", "초코크런치/스프링클"],
                        ["젓가락핫케이크가루", "핫케이크가루"],
                        ["다진쇠고기순두부", "다진쇠고기/순두부"],
                        ["멸치다시마육수", "멸치/다시마육수"],
                        ["애호박청양고추", "애호박/청양고추"],
                        ["우엉카레가루", "우엉/카레가루"],
                        ["박력분버터", "박력분/버터"],
                        ["대파명랏젓", "대파/명랏젓"],
                        ["오이양파", "오이/양파"],
                        ["설탕소금", "설탕/소금"],
                        ["맛술미림", "맛술/미림"],
                        ["미숫가루막걸리막걸리", "미숫가루/막걸리"],
                        ["후춧가루다진마늘", "후춧가루/다진마늘"],
                        ["설탕달걀노른자", "설탕/달걀"],
                        ["다진마늘참기름", "다진마늘/참기름"],
                        ["설탕포도씨유", "설탕/포도씨유"],
                        ["소금올리브유", "소금/올리브유"],
                        ["밀가루오징어", "밀가루/오징어"],
                        ["국간장소금", "국간장/소금"],
                        ["주키니호박다진돼지고기", "주키니호박/다진돼지고기"],
                        ["설탕단호박라떼단호박", "설탕/단호박"],
                        ["식초홀그레인머스터드", "식초/홀그레인머스터드"],
                        ["통조림골뱅이대파", "통조림골뱅이/대파"],
                        ["다진마늘참기름", "다진마늘/참기름"],
                        ["다진생강소금", "다진생강/소금"],
                        ["비트채나새싹", "비트/새싹"],
                        ["다진마늘식초", "다진마늘/식초"],
                        ["한라봉버터", "한라봉/버터"],
                        ["참기름참깨", "참기름/참깨"],
                        ["설탕검은깨", "설탕/검은깨"],
                        ["대파물만두", "대파/물만두"],
                        ["우유파르메산치즈간것", "우유/파르메산치즈간것"],
                        ["베이컨슈레드모차렐라치즈", "베이컨/슈레드모차렐라치즈"],
                        ["김치부침가루", "김치/부침가루"],
                        ["젓가락핫케이크가루", "핫케이크가루"],
                        ["두부감자", "두부/감자"]]

mixed_ingredients = []
for mixed_ingredient in mixed_ingredient_list:
    mixed_ingredients.append(mixed_ingredient[0])

for i in range(len(data['data'])):
    for c in range(len(data['data'][i]['ingredients'])):
        ingredient = data['data'][i]['ingredients'][c]

        if(ingredient in mixed_ingredients):
            for mixed_ingredient in mixed_ingredient_list:
                if(ingredient == mixed_ingredient[0]):
                    seperate = mixed_ingredient[1]
            for k in seperate.split("/"):
                data['data'][i]['ingredients'].append(k)
            del(data['data'][i]['ingredients'][c])
```

```python
### 4. 불필요한 접두어, 접미사 제거

# 접두어
prefix_list = ["시판", "선택재료", "불린", "김밥용", "편의점", "삶은", 
                "실온에꺼내둔", "으깬", "완숙", "무가당", "통", "껍질깐", 
                "껍질벗긴", "비정제", "데친", "감칠맛", "고운", "얼린",
                "생식용", "밑간", "해감", "볶은", "부순", "다진", "부각용", 
                "삭힌", "손질"]

for i in range(len(data['data'])):
    for c in range(len(data['data'][i]['ingredients'])):
        ingredient = data['data'][i]['ingredients'][c]

        for prefix in prefix_list:
            if(prefix in ingredient):

                # 통조림 예외처리
                if(prefix == "통"):
                    if(ingredient[:3] != "통조림"):
                        data['data'][i]['ingredients'][c] = ingredient.split(prefix)[1]
                
                # 다진마늘, 다진소고기, 다진돼지고기 예외처리
                elif(prefix == "다진"):
                    if(ingredient[2:] not in ["소고기", "쇠고기", "마늘", "돼지고기"]):
                        data['data'][i]['ingredients'][c] = ingredient.split(prefix)[1]
                
                # 손질된, 손질한 처리
                elif(prefix == "손질"):
                    if(ingredient[:3] == "손질된"):
                        data['data'][i]['ingredients'][c] = ingredient.split("손질된")[1]
                    elif(ingredient[:3] == "손질한"):
                        data['data'][i]['ingredients'][c] = ingredient.split("손질한")[1]
                    else:
                        data['data'][i]['ingredients'][c] = ingredient.split(prefix)[1]

                else:
                    data['data'][i]['ingredients'][c] = ingredient.split(prefix)[1]

# 접미사
suffix_list = ["간것", "우린물", "푸른부분", "밥", "색소"]
for i in range(len(data['data'])):
    del_idx = []
    for c in range(len(data['data'][i]['ingredients'])):
        ingredient = data['data'][i]['ingredients'][c]

        for suffix in suffix_list:
            if(suffix in ingredient):
                # 모든 밥은 즉석밥으로 변경
                if(suffix == "밥"):
                    # 김밥, 삼각김밥 예외처리
                    if(ingredient[-2:] != "김밥"):
                        data['data'][i]['ingredients'][c] = "즉석밥"

                # 돼지고기간것에 대한 예외처리 => 다진돼지고기
                elif(suffix == "간것"):
                    if(ingredient =="돼지고기간것"):
                        data['data'][i]['ingredients'][c] = "다진돼지고기"
                    else:
                        data['data'][i]['ingredients'][c] = ingredient.split(suffix)[0]
                # 색소가 들어간 재료는 삭제
                elif(suffix == "색소"):
                    del_idx.append(c)
                else:
                    data['data'][i]['ingredients'][c] = ingredient.split(suffix)[0]
    del_idx.sort()
    # 저장된 인덱스의 재료를 삭제
    cnt = 0
    for idx in del_idx:
        idx = idx - cnt
        del(data['data'][i]['ingredients'][idx])
        # 앞에 재료가 하나 삭제 되면 인덱스가 하나씩 밀리게 되므로 처리
        cnt = cnt + 1

# 중간에 포함되어 있는 것
mid_list = ["썬", "만두"]
for i in range(len(data['data'])):
    for c in range(len(data['data'][i]['ingredients'])):
        ingredient = data['data'][i]['ingredients'][c]

        for mid in mid_list:
            if(mid in ingredient):
                # ~썬- 에서 썬을포함한 앞부분 제거
                if(mid == "썬"):
                    data['data'][i]['ingredients'][c] = ingredient.split(mid)[1]
                
                # 만두 중에서 냉동만두를 일반 만두로 변경
                elif(mid == "만두"):
                    if("냉동" in ingredient):
                        data['data'][i]['ingredients'][c] = ingredient.split("냉동")[1]


                
```

```python
### 5. 유사/동일재료 이름 통합

replace_list = [   
    ["소고기/쇠고기", "소고기"],
    ["비네거/식초", "식초"],
    ["가루/파우더", "파우더"],
    ["달걀/계란", "계란"],
    ["붉은양파/적양파", "적양파"],
    ["붉은색파프리카/빨간파프리카/붉은파프리카", "빨간파프리카"],
    ["노란색파프리카/노란파프리카", "노란파프리카"],
    ["말린/마른/건", "건"],
    ["피망/청피망", "청피망"],
    ["홍피망/붉은피망", "홍피망"],
    ["파르메산치즈/덩어리파르메산치즈", "파르메산치즈"],
    ["모차렐라치즈/프레시모차렐라/덩어리연질치즈/프레시모차렐라치즈/생모차렐라/모차렐라", "모차렐라치즈"],
    ["김/김밥용김/마른김/구운김/파래김/일반김/곱창김/돌김/재래김", "김"],
    ["김가루/조미김가루/파래김가루", "김가루"],
    ["조미김/조미재래김", "조미김"],
    ["김치/배추김치/잘게썬배추김치/송송썬김치/다진김치/익은김치/잘게썬김치", "김치"],
    ["단무지/원형단무지", "단무지"],
    ["게맛살/맛살/꽃맛살", "맛살"],
    ["우엉조림/조림우엉", "우엉조림"],
    ["국물용멸치/국물용손질멸치/육수용멸치", "국물용멸치"],
    ["아삭이고추/오이고추", "오이고추"],
    ["베트남고추/태국고추", "베트남고추"],
    ["고추/청고추", "청고추"],
    ["붉은고추/홍고추", "홍고추"],
    ["와사비/고추냉이", "고추냉이"],
    ["생와사비/생고추냉이/생고추냉이가루", "생고추냉이"],
    ["연와사비/연고추냉이", "연고추냉이"],
    ["보해맛술/맛술", "맛술"],
    ["닭육수/닭고기육수", "닭육수"],
    ["표고버섯육수/마른표고버섯육수/표고버섯국물", "표고버섯"],
    ["채소육수/채소국물", "야채스톡"],
    ["장조림고기", "장조림"],
    ["새우젓국물", "새우젓"],
    ["사각어묵/어묵/판어묵", "사각어묵"],
    ["막대어묵/꼬치어묵", "꼬치어묵"],
    ["봉어묵/구멍봉어묵", "봉어묵"],
    ["취향에맞는치즈", "모차렐라치즈"],
    ["브로콜리줄기/브로콜리", "브로콜리"],
    ["굵게다진마늘/다진마늘", "다진마늘"],
    ["다크초콜릿/제과용다크초콜릿/코팅용다크초콜릿/다크커버춰초콜릿", "다크초콜릿"],
    ["코팅용화이트초콜릿/화이트초콜릿/화이트초콜릿커버쳐", "화이트초콜릿"],
    ["후춧가루/백후춧가루/크러쉬드페퍼", "후춧가루"],
    ["찜용콩나물/콩나물찜용/콩나물", "콩나물"],
    ["연어/생연어/연어살/횟감용연어/연어필레", "생연어"],
    ["빙수용팥/팥/설탕에조린팥/조림팥", "팥"],
    ["주머니곶감/곶감", "곶감"],
    ["파슬리가루/파슬리/생파슬리", "파슬리"],
    ["이탈리안파슬리/다진이탈리안파슬리", "이탈리안파슬리"],
    ["재래된장/된장", "된장"],
    ["홀토마토/토마토/완숙토마토", "토마토"],
    ["바닐라아이스크림/바닐라맛아이스크림", "바닐라아이스크림"],
    ["마른보리새", "보리새우"],
    ["취향에맞는치즈", "모차렐라치즈"],
    ["물/미지근한물/쌀뜨물/따뜻한물/찬물/얼음물", "물"],
    ["달걀물", "계란"],
    ["다시마물", "다시마"],
    ["전분물", "전분"]
]

for i in range(len(data['data'])):
    for c in range(len(data['data'][i]['ingredients'])):
        ingredient = data['data'][i]['ingredients'][c]

        for replace in replace_list:
            temp_list = replace[0].split("/")
            for temp in temp_list:
                if(ingredient == temp):
                    data['data'][i]['ingredients'][c] = replace[1]
```
