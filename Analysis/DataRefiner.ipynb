```python
import json
import pandas as pd
```

```python
with open('recipe_data.json', encoding="utf-8") as f:
    data = json.load(f)
```

```python
data
```

```python
lst = []
for i in range(len(data['data'])):
    lst.append(data['data'][i]['ingredients'])
```

```python
lst
```

```python
ingred = pd.DataFrame({"ingredients":lst})
```

```python
detail = []
for i in ingred["ingredients"]:
    for j in i:
        if ((j != '') and ("요." not in j)):
            if ("TIP" in j):
                j = (j.split("TIP")[0])
            elif ("tip" in j):
                j = (j.split("tip")[0])
            elif ("Tip" in j):
                j = (j.split("Tip")[0])
                
            if ("(" in j):
                j = (j.split("(")[0])
            elif (")" in j):
                j = (j.replace(")",""))
            
            if ("+" in j):
                for k in j.split("+"):
                    detail.append(k)
            elif ("또는" in j):
                for k in j.split("또는"):
                    detail.append(k)
            elif ("." in j):
                for k in j.split("."):
                    detail.append(k)
            else :
                detail.append(j)
```

```python
detail
```

```python
sum = pd.DataFrame({"total":detail})
sum.value_counts(ascending=False)
```

```python
sum.value_counts(ascending=False).to_excel("countIngredients.xlsx")
```
