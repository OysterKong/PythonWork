```python
# 문제 1번

# 1) 1차 행 인덱스로 "반"을 2차 행 인덱스로 "번호"를 가지는 데이터 프레임 작성

# 2) 학생의 평균을 나타내는 행을 오른쪽에 추가

df_score1 = pd.DataFrame({
    "반":["A", "A", "B", "A", "B", "B", "B", "A", "B", "A"],
    "번호" : [1, 2, 1, 3, 2, 3, 4, 4, 5, 5],
    "국어" : [79, 67, 88, 68, 92, 54, 67, 88, 97, 85],
    "영어" : [55, 77, 44, 67, 86, 45, 78, 58, 90, 67],
    "수학" : [57, 45, 76, 68, 89, 67, 99, 78, 89, 90]
})

df_score2=df_score1.sort_values(by='반')
df_score3=df_score2.set_index(["반","번호"])
df_score3

df_score3["평균"] = round(df_score3.mean(axis=1),2)
df_score3
```

```python
# 문제 2번

# 1) 행 인덱스로 "번호"를 1차 열 인덱스로 "국어", "영어", "수학"을, 2차 열 인덱스로 "반"을 가지는 데이터 프레임 작성

# 2) 각 반별 각 과목의 평균을 나타내는 행을 아래에 추가 

df_score4 =df_score3.unstack("반").sort_index(level=0)
df_score4

df_score4.loc["평균",:] = df_score4.mean()
df_score4
```

```python
# 문제 5번

bike = pd.read_csv("data/SeoulBikeData.csv")

# 1) weekdays 파생변수를 만들고 "Date" 컬럼에서 날짜를 추출하여 요일로 변환한 뒤 파생변수에 저장하시오.

bike["Date"]=pd.to_datetime(bike["Date"], format="%d/%m/%Y")
bike.info()
diff=bike["Date"] - pd.to_datetime("2016-1-1")
diff=diff.astype(str)
diff_num=diff.str.split(" ", expand=True)[0].astype(int) #expand=False : 기본값

def weekdays(x):
    if x %7 == 0:
        return "금요일"
    elif x % 7 ==1:
        return "토요일"
    elif x % 7 ==2:
        return "일요일"
    elif x % 7 ==3:
        return "월요일"
    elif x % 7 ==4:
        return "화요일"
    elif x % 7 ==5:
        return "수요일"
    elif x % 7 ==6:
        return "목요일"
    
bike["weekdays"] = diff_num.apply(weekdays)
bike.head()
```

```python
# 2) 요일별로 시간대별 그래프로 시각화 하시오

_,axes = plt.subplots(2,4, figsize=(20,10))
w_list=["월요일","화요일","수요일","목요일","금요일","토요일","일요일"]
for i in range(7):
    sns.lineplot("Hour","Rented Bike Count", 
                 data=bike.loc[bike["weekdays"]==w_list[i]], ax = axes[i//4, i%4])
```
