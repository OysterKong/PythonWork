# 1. 커널 기법

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(centers=4, random_state=8)
y = y%2

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
```

```python
linear_svm = LinearSVC().fit(X, y)

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
mglearn.plots.plot_2d_separator(linear_svm, X)
```

```python
X_new = np.hstack([X, X[:, 1:] ** 2])

from mpl_toolkits.mplot3d import Axes3D, axes3d
figure = plt.figure()
# 3차원 그래프
ax = Axes3D(figure, elev=-152, azim=-26)

mask = y == 0
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b', cmap=mglearn.cm2, s=60,  edgecolor='k')
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', marker='^', 
           cmap=mglearn.cm2, s=60, edgecolor='k')
ax.set_xlabel("특성0")
ax.set_ylabel("특성1")
ax.set_zlabel("특성1 ** 2")
```

```python
linear_svm_3d = LinearSVC().fit(X_new, y)
coef, intercept = linear_svm_3d.coef_.ravel(), linear_svm_3d.intercept_
# 선형 결정 경계 그리기
figure = plt.figure()
ax = Axes3D(figure, elev=-152, azim=-26)
xx = np.linspace(X_new[:, 0].min() - 2, X_new[:, 0].max() + 2, 50)
yy = np.linspace(X_new[:, 1].min() - 2, X_new[:, 1].max() + 2, 50)

XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]
ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=0.3)
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b', cmap=mglearn.cm2, s=60,  edgecolor='k')
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', marker='^', 
           cmap=mglearn.cm2, s=60, edgecolor='k')
ax.set_xlabel("특성0")
ax.set_ylabel("특성1")
ax.set_zlabel("특성1 ** 2")
```

```python
ZZ = YY ** 2
dec = linear_svm_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])
plt.contourf(XX, YY, dec.reshape(XX.shape), levels=[dec.min(), 0, dec.max()], 
            cmap=mglearn.cm2, alpha=0.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

### (5) Pipeline

```python
import multiprocessing
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)
```

#### 1) 첫번째 방법

```python
pipe = [("scaler", StandardScaler()), ("svm", SVC())]   # 내가 해야할 작업을 순서대로

pipeline = Pipeline(pipe, verbose=True)     # Pipeline 이라는 객체가 pipe에 담긴 작업을 순서대로 진행할 수 있게끔 해준다.
pipeline.steps
```

```python
pipeline.fit(X_train, y_train)

pred_train = pipeline.predict(X_train)  # 훈련결과 저장
pred_test = pipeline.predict(X_test)   # 실제테스트 결과 저장

print(accuracy_score(y_train, pred_train))
print(accuracy_score(y_test, pred_test))
```

#### 2) 두번째 방법

```python
pipeline = make_pipeline(StandardScaler(), SVC())

pipeline.fit(X_train, y_train)

pred_train = pipeline.predict(X_train)  # 훈련결과 저장
pred_test = pipeline.predict(X_test)   # 실제테스트 결과 저장

print(accuracy_score(y_train, pred_train))
print(accuracy_score(y_test, pred_test))
```

#### 세번쨰 방법 : GridSearchCV를 이용하는 방법

```python
pipe = [("scaler", StandardScaler()), ("svm", SVC(kernel="linear"))]

pipeline = Pipeline(pipe, verbose=True)

params = {"svm__C":[0.01, 0.1, 1, 10, 100], "svm__gamma":[0.01, 0.1, 1, 10, 100]}
gsc = GridSearchCV(pipeline, param_grid=params, cv=5)

gsc.fit(X_train, y_train)
#pd.DataFrame(gsc.cv_results_)

print(gsc.score(X_train, y_train))
print(gsc.score(X_test, y_test))
```

#### 4) 연습

```python
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = SVC()

params = {"C":[1.0, 0.1, 0.01, 10], "gamma":["scale", "auto"], "kernel":["rbf", "poly", "linear", "sigmoid"]}
gs = GridSearchCV(estimator=model, param_grid=params, n_jobs=multiprocessing.cpu_count(), cv=5, verbose=True)
gs.fit(X_train, y_train)

print(gs.score(X_train, y_train))
print(gs.score(X_test, y_test))
```

```python
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)

pipe = [("scaler", StandardScaler()), ("model", SVC())]
pipeline = Pipeline(pipe, verbose=True)

params = {"model__C":[1.0, 0.1, 0.01, 10], "model__gamma":["scale", "auto"], "model__kernel":["rbf", "poly", "linear", "sigmoid"]}
gs = GridSearchCV(estimator=pipeline, param_grid=params, n_jobs=multiprocessing.cpu_count(), cv=5, verbose=True)
gs.fit(X_train, y_train)

print(gs.score(X_train, y_train))
print(gs.score(X_test, y_test))
```

### (6) 손글씨 이미지(MNIST)

```python
def load_csv(fname):
    labels = []
    images = []
    
    f = open(fname, "r")
    for line in f:
        cols = line.split(",")
        
        if len(cols) < 2:
            continue
            
        labels.append(int(cols.pop(0)))
        
        vals = list(map(lambda n : int(n)/256, cols))
        images.append(vals)
    
    f.close()
    return {"y":labels, "X":images}
```

```python
data = load_csv("data/MNIST_data/train.csv")
test = load_csv("data/MNIST_data/test.csv")
```
